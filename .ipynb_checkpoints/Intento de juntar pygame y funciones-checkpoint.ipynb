```python
# Creamos un array bidimensional. Un array bidimensional
# no es más que una lista de listas.
board_player = [[6 for slot in range(4)] for row in range(12)]
board_response = [[6 for slot in range(4)] for row in range(12)]
color_options = [0,1,2,3,4,5,6]
code = [1,2,3,4]
grid = [[0 for slot in range(4)] for row in range(12)]

import random

def computer_choice():
    return random.choices(range(0,6), k=4)

def player_choice():
    return [int(i) for i in list(input("Insert 4 numbers between 0 and 5 (you can repeat numbers): "))]

def comparing(a, b):
    for i, j in zip(a, b):
        if i == j:
            cpu_response.append("B")
        elif i != j:
            if i in b: 
                cpu_response.append("W")
            else:
                cpu_response.append("X")
    return cpu_response


"""
 Programa de ejemplo de cómo usar un array para respaldar una retícula en pantalla.
  
 Sample Python/Pygame Programs
 Simpson College Computer Science
 http://programarcadegames.com/
 http://simpson.edu/computer-science/
 
Vídeo explicativo: http://youtu.be/mdTeqiWyFnc
"""
import pygame
 
# Definimos algunos colores
NEGRO = (0, 0, 0)
BLANCO = (255, 255, 255)
AZUL = (0, 0, 255)
VERDE = (0, 255, 0)
ROJO = (255, 0, 0)
AMARILLO = (255, 255, 0)
MORADO = (255, 0, 255)
NARANJA = (255, 120, 0)
 
# Establecemos el LARGO y ALTO de cada celda de la retícula.
LARGO  = 10
ALTO = 10
 
# Establecemos el margen entre las celdas.
MARGEN = 10
 
 
# Establecemos la fila 1, celda 5 a uno. (Recuerda, los números de las filas y
# columnas empiezan en cero.)
grid[0][0] = 1
 
# Inicializamos pygame
pygame.init()
  
# Establecemos el LARGO y ALTO de la pantalla
DIMENSION_VENTANA = [190, 320]
pantalla = pygame.display.set_mode(DIMENSION_VENTANA)
 
# Establecemos el título de la pantalla.
pygame.display.set_caption("Retículas y Matrices")
 
# Iteramos hasta que el usuario pulse el botón de salir.
hecho = False
 
# Lo usamos para establecer cuán rápido de refresca la pantalla.
reloj = pygame.time.Clock()
 
# -------- Bucle Principal del Programa-----------
while not hecho:
    for evento in pygame.event.get(): 
        if evento.type == pygame.QUIT: 
            hecho = True
        elif evento.type == pygame.MOUSEBUTTONDOWN:
            # El usuario presiona el ratón. Obtiene su posición.
            pos = pygame.mouse.get_pos()
            # Cambia las coordenadas x/y de la pantalla por coordenadas reticulares
            columna = pos[0] // (LARGO + MARGEN)
            fila = pos[1] // (ALTO + MARGEN)
            # Establece esa ubicación a cero
            grid[fila][columna] = 1
            print("Click ", pos, "Coordenadas de la retícula: ", fila, columna)
 
    # Establecemos el fondo de pantalla.
    pantalla.fill(NEGRO)
 
    # Dibujamos la retícula del board_player
    for row in range(12):
        for slot in range(4):
            color = BLANCO
            if board_player[row][slot] == 0:
                color = ROJO
            elif board_player[row][slot] == 1:
                color = NARANJA
            elif board_player[row][slot] == 2:
                color = AZUL 
            elif board_player[row][slot] == 3:
                color = AMARILLO 
            elif board_player[row][slot] == 4:
                color = MORADO
            elif board_player[row][slot] == 5:
                color = VERDE 
            elif board_player[row][slot] == 6:
                color = BLANCO 
            elif board_player[row][slot] == 7:
                color = NEGRO 

            pygame.draw.rect(pantalla,
                             color,
                             [(MARGEN+LARGO) * slot + MARGEN,
                              (MARGEN+ALTO) * row + MARGEN + 40,
                              LARGO,
                              ALTO])
            
    # Dibujamos la retícula del board_response
    for row in range(12):
        for slot in range(4):
            color = BLANCO
            if board_response[row][slot] == 0:
                color = ROJO
            elif board_response[row][slot] == 6:
                color = BLANCO 
            elif board_response[row][slot] == 7:
                color = NEGRO 

            pygame.draw.rect(pantalla,
                             color,
                             [(MARGEN+LARGO) * slot + MARGEN + 100,
                              (MARGEN+ALTO) * row + MARGEN + 40,
                              LARGO,
                              ALTO])
            
#    # Dibujamos la retícula del color_options
#    for fila in range(1):
#        for columna in range(6):
#            color = BLANCO
#            if color_options[x] == 1:
#                color = VERDE 
#            buttons = pygame.draw.rect(pantalla,
#                             color,
#                             [(MARGEN+LARGO) * columna + MARGEN + 30,
#                              (MARGEN+ALTO) * fila + MARGEN + 10,
#                              LARGO,
#                              ALTO])

    # Dibujamos la retícula del color_options
    for elem in color_options[1:]:
            color = BLANCO
            if color_options[elem] == 0:
                color = ROJO
            elif color_options[elem] == 1:
                color = NARANJA
            elif color_options[elem] == 2:
                color = AZUL 
            elif color_options[elem] == 3:
                color = AMARILLO 
            elif color_options[elem] == 4:
                color = MORADO
            elif color_options[elem] == 5:
                color = VERDE 
            elif color_options[elem] == 6:
                color = BLANCO 
            elif color_options[elem] == 7:
                color = NEGRO 
            pygame.draw.rect(pantalla,
                             color,
                             [(MARGEN+LARGO) * elem + MARGEN +10,
                              (MARGEN+ALTO) * 1 + MARGEN,
                              LARGO,
                              ALTO])




    # Dibujamos la retícula del CODE:!!!!!!!!!!!!!!!!!
    for slot in range(6):
        color = BLANCO
        if code[elem] == 0:
            color = ROJO 
        elif code[elem] == 1:
            color = NARANJA
        elif code[elem] == 2:
            color = AZUL 
        elif code[elem] == 3:
            color = AMARILLO 
        elif code[elem] == 4:
            color = MORADO
        elif code[elem] == 5:
            color = VERDE 
              
        pygame.draw.rect(pantalla,
                             color,
                             [(MARGEN+LARGO) * slot + MARGEN + 30,
                              (MARGEN+ALTO) * 1 + MARGEN + 260,
                              LARGO,
                              ALTO])

            
    # Limitamos a 60 fotogramas por segundo.
    reloj.tick(60)
 
    # Avanzamos y actualizamos la pantalla con lo que hemos dibujado.
    pygame.display.flip()
     
# Pórtate bien con el IDLE.
pygame.quit()
```
